# Notion de cours :

Il existe plusieurs types d'adresses IP :

- De type A allant de 0.0.0.0 à 127.255.255.255 
- De type B allant de 128.0.0.0 à 191.255.255.255
- De type C allant de 192.0.0.0 à 223.255.255.255

Chaque adresse possède son masque réseau :

- Type A : Le masque réseau est 255.0.0.0
- Type B : Le masque réseau est 255.255.0.0
- Type C : Le masque réseau est 255.255.255.0

Un réseaux possede deux adresse "spéciales" **l'adresse de broadcast** (de diffusion) et **l'adresse réseaux**. 


Comment determiner l'adresse réseaux et l'adresse de broadcast ? 

Voici une adresse IPV4 : 192.155.255.42.
De type C son masque réseaux est 255.255.255.0


### L'adresse réseaux :

**L'addresse réseaux** se détermine par une opération ET bit à bit entre l'adresse IP et son masque réseaux ici 192.155.255.42 et 255.255.255.0

L'adresse réseaux est donc : **192.155.255.0**

### L'adresse de broadcast : 

**L'adresse de broadcast** se détermine à partir de l'adresse réseaux et du masque réseaux **inversé** (ou sous réseaux)
On fait un OU bit à bit entre l'adresse réseaux et le masque réseaux **inversé** 

L'adresse inversée de 255.255.255.0 est 0.0.0.255. 

De manière générale on **inverse** une adresse en passant les 0 en 1 et les 1 en 0 (11111111.11111111.11111111.00000000 => 00000000.00000000.00000000.11111111) 

Il faut donc faire un OU bit à bit avec 192.155.255.0 et 0.0.0.255

On obtient donc :  192.155.255.255

L'adresse de broadcast de l'adresse 192.155.255.42 est **192.155.255.255**


#### Remarque : Les adresse IP sont contenuent entre l'adresse réseaux et l'adresse de broadcast.

Ici les adresse IP disponibles pour ce réseau sont comprises entre 192.155.255.0 et 192.155.255.255

Les sous réseaux sont utilisés pour structuré notre réseaux pour ce faire nous allons attribuer une partie de notre adresse réseaux au réseaux et l'autre pour les hôtes de chaques sous réseaux.

Par exemple : 

Prennons l'adresse réseaux : 192.155.255.0 
Son masque réseau est 255.255.255.0 (Classe C)

Le masque de sous réseau lui commencera forcément par **255.255.255.xx** 

Supposons un masque de sous réseaux 255.255.255.224 et une adresse 192.155.255.33 

### Savoir à quel sous réseaux l'adresse ip appartient elle :
 
Il suffit de faire une opération bit à bit entre le masque de sous réseaux et l'adresse ip 

Ici 192.155.255.33 ET 255.255.255.224 soit : 192.155.255.32 

L'adresse IP 192.155.255.33 appartient au sous réseau 192.155.255.32

### Comment attribuer des sous réseaux à une adresse réseaux ?

Dans notre exemple j'ai 24 bits attribués à mon masque de réseaux, il m'en reste donc 8 pour créer un masque sous réseaux.

Je décide d'attribuer par exemple 3 bits pour l'adressage de sous réseaux, il m'en reste donc 5 pour l'hôte.

Mon masque de sous réseaux devient 11111111.11111111.11111111.11100000 soit 255.255.255.224

En ayant attribué 3 bits pour les sous réseaux j'ai donc en théorie 8 sous réseaux 2**3 = 8. 

Soit :
    - 255.255.255.000 00000 => 255.255.255.0
    - 255.255.255.001 00000 => 255.255.255.32
    - 255.255.255.010 00000 => 255.255.255.64
    - 255.255.255.011 00000 => 255.255.255.96
    - 255.255.255.100 00000 => 255.255.255.128
    - 255.255.255.101 00000 => 255.255.255.160
    - 255.255.255.110 00000 => 255.255.255.192
    - 255.255.255.111 00000 => 255.255.255.224
    
Sauf que l'adresse de sous réseau 255.255.255.0 est équivalente à l'adresse réseaux 255.255.255.0 donc elle ne sera pas utilisée
